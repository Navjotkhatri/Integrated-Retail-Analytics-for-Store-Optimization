## Dataset Understanding

The project uses three datasets: Sales Data, Features Data, and Store Data. These datasets are integrated to perform retail analytics, demand forecasting, anomaly detection, and store optimization.

### 1. Sales Dataset

**Columns:**

**Purpose:**
This dataset helps analyze sales trends, identify unusual sales patterns, and build machine learning models to forecast future demand.


### 2. Features Dataset

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization (optional for checking)
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
````

````python
f_df['Date'] = pd.to_datetime(
    f_df['Date'],
    format='%d/%m/%Y'
)

sales_df['Date'] = pd.to_datetime(
    sales_df['Date'],
    format='%d/%m/%Y'
)
````

````python
markdown_cols = [
    'MarkDown1',
    'MarkDown2',
    'MarkDown3',
    'MarkDown4',
    'MarkDown5'
]

f_df[markdown_cols] = f_df[markdown_cols].fillna(0)
````

````python
f_df['CPI'] = f_df['CPI'].interpolate()

f_df['Unemployment'] = f_df['Unemployment'].interpolate()
````

````python
f_df['Total_MarkDown'] = (
    f_df['MarkDown1'] +
    f_df['MarkDown2'] +
    f_df['MarkDown3'] +
    f_df['MarkDown4'] +
    f_df['MarkDown5']
)
````

````python
f_df['Has_MarkDown'] = np.where(
    f_df['Total_MarkDown'] > 0,
    1,
    0
)
````

````python
f_df['Year'] = f_df['Date'].dt.year
f_df['Month'] = f_df['Date'].dt.month
f_df['Week'] = f_df['Date'].dt.isocalendar().week

sales_df['Year'] = sales_df['Date'].dt.year
sales_df['Month'] = sales_df['Date'].dt.month
sales_df['Week'] = sales_df['Date'].dt.isocalendar().week
````

````python
f_df['IsHoliday'] = f_df['IsHoliday'].astype(int)

sales_df['IsHoliday'] = sales_df['IsHoliday'].astype(int)
````

````python
sales_features_df = sales_df.merge(
    f_df,
    on=['Store','Date'],
    how='left'
)

sales_features_df.head()
````

````python
retail_master_df = sales_features_df.merge(
    store_df,
    on='Store',
    how='left'
)

retail_master_df.head()
````

````python
print("Final Dataset Shape:")
print(retail_master_df.shape)

retail_master_df.info()
````

````python
retail_master_df.isnull().sum()
````

````python
retail_master_df.to_csv(
    'retail_master_cleaned.csv',
    index=False
)
````

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Do holiday weeks generate significantly different sales compared to non-holiday weeks?

Null Hypothesis (H₀):

There is no significant difference in average weekly sales between holiday and non-holiday weeks.

Alternative Hypothesis (H₁):

There is a significant difference in average weekly sales between holiday and non-holiday weeks.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import ttest_ind

holiday_sales = retail_master_df[
    retail_master_df['IsHoliday_x'] == 1
]['Weekly_Sales']

non_holiday_sales = retail_master_df[
    retail_master_df['IsHoliday_x'] == 0
]['Weekly_Sales']


t_stat, p_value = ttest_ind(
    holiday_sales,
    non_holiday_sales,
    equal_var=False
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
````

    Fail to reject H0

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = [
    'Weekly_Sales',
    'Temperature',
    'Fuel_Price',
    'CPI',
    'Unemployment',
    'Total_MarkDown',
    'Size'
]

plt.figure(figsize=(18,10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3,3,i)
    sns.boxplot(y=retail_master_df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
````

````python
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]

    return len(outliers)

for col in numeric_cols:
    print(f"{col}: {count_outliers(retail_master_df, col)} outliers")
````

````python
def cap_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[column] = df[column].clip(lower, upper)

    return df

cols_to_cap = [
    'Weekly_Sales',
    'Temperature',
    'Fuel_Price',
    'CPI',
    'Unemployment',
    'Total_MarkDown'
]

for col in cols_to_cap:
    retail_master_df = cap_outliers(retail_master_df, col)
````

````python
plt.figure(figsize=(18,10))

for i, col in enumerate(cols_to_cap, 1):
    plt.subplot(2,3,i)
    sns.boxplot(y=retail_master_df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outlier treatment was performed to reduce the impact of extreme values on data analysis and machine learning models while preserving meaningful business information.

Instead of deleting outliers, Winsorization (capping) was applied. Values below the lower bound were replaced with the lower limit, and values above the upper bound were replaced with the upper limit.


The combination of IQR-based detection and Winsorization (capping) was chosen because it effectively handles extreme values while maintaining the integrity of the retail sales data. This approach improves the reliability of statistical analysis and predictive models without losing important information related to seasonal demand, promotions, and holiday sales.

## ***7. ML Model Implementation***

### ML Model - 1 K-Means Clustering (Store Segmentation)

````python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Aggregate store-level data
store_cluster = retail_master_df.groupby('Store').agg({
    'Weekly_Sales':'mean',
    'Total_MarkDown':'mean',
    'Temperature':'mean',
    'Fuel_Price':'mean',
    'CPI':'mean',
    'Unemployment':'mean',
    'Size':'first'
}).reset_index()

# Scale features
X_cluster = store_cluster.drop('Store', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Train KMeans
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

store_cluster['Cluster'] = kmeans.fit_predict(X_scaled)

store_cluster.head()
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, store_cluster['Cluster'])

print("Silhouette Score:", score)
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in range(2, 11):
    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = model.fit_predict(X_scaled)

    silhouette_scores.append(
        silhouette_score(X_scaled, labels)
    )

for k, score in zip(range(2,11), silhouette_scores):
    print(f"K={k}, Silhouette Score={score:.3f}")
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(range(2,11), silhouette_scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Selecting Optimal K")
plt.grid(True)
plt.show()
````

The optimal number of clusters was selected using the Silhouette Score. Since K-Means is an unsupervised algorithm, traditional cross-validation was not applicable.
